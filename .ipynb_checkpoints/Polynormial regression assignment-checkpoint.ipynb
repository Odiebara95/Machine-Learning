```python
from sklearn import datasets
```

```python
boston=datasets.load_boston()
```

```python
print(boston.DESCR)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=boston.data
y=boston.target
```

```python
X_room=X[:,5]
X_room
```

```python
from sklearn.linear_model import LinearRegression
```

```python
plt.scatter(X_room,y)
```

```python
X_room=X_room.reshape(-1,1)
```

```python
linear_model =LinearRegression()
linear_model.fit(X_room,y)
```

```python
linear_model.score(X_room,y)
```

```python
y_pred=linear_model.predict(X_room)
plt.scatter(X_room,y,alpha=0.5,c='r')
plt.plot(X_room,y_pred)
```

## polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
room_poly1=PolynomialFeatures(include_bias=False)
```

```python
room_poly1.fit(X_room)
```

```python
poly_Xroom=room_poly1.transform(X_room)
```

```python
model1=LinearRegression()
```

```python
model1.fit(poly_Xroom,y)
y_pred1=model1.predict(poly_Xroom)
```

```python
plt.scatter(X_room.ravel(),y)
plt.plot(X_room.ravel(),y_pred1,c="r")
```

## degree 3

```python
room_poly3=PolynomialFeatures(degree=3,include_bias=False)
room_poly3.fit(X_room)
poly_Xroom3=room_poly3.transform(X_room)
model3=LinearRegression()
model3.fit(poly_Xroom3,y)
y_pred3=model3.predict(poly_Xroom3)
```

```python
plt.scatter(X_room.ravel(),y)
plt.plot(X_room.ravel(),y_pred3,c="r")
```
